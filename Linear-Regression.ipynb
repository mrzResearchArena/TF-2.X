```
D = pd.read_csv('moore.csv', header=None).values
X = D[:,0].reshape(-1, 1)
Y = D[:,1]
```

```
X = X - X.mean()
```

```
plt.scatter(X, Y)
```

```
Y = np.log(Y)
plt.scatter(X, Y)
```

```
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)                                    
])
```

```
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.001), # momentum=?
    loss='mse',
)
```

```
# scheduler = tf.keras.callbacks.LearningRateScheduler(lrSchedule)
```

```
result = model.fit(X, Y, epochs=500,) # callbacks=[scheduler]
```
