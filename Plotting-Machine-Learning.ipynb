```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.arange(0, 10, 0.25) # 0<10, increase by 1 (0, 0.25, 0.5, 0.75, ... 9.75)
x
```

### [1]. Drawing a General Plot:

```
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()

plt.title('General Plot')
plt.xlabel('Time')
plt.ylabel('Value')

plt.show()
# end
```

### [2]. Drawing a Sclatter Plot:

```
plt.scatter(x, np.sin(x), label='sin')
plt.scatter(x, np.cos(x), label='cos')
plt.legend()

plt.title('Sclatter Plot')
plt.xlabel('Time')
plt.ylabel('Value')

plt.show()
# end
```

### [3]. Drawing a General Plot and Sclatter Plot Together:

```
plt.scatter(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()

plt.title('Scaller Plot and General Plot Together')
plt.xlabel('Time')
plt.ylabel('Value')

plt.show()
# end
```

### [4]. Bar Plot / Frequency Plot:

```
names  = ['A', 'C', 'G', 'T']
values = [10, 5, 14, 11] 

plt.bar(names, values)
plt.title('Frequency Plot')

plt.show()
# end
```

### [5]. Pie Plot:

```
# Using above cell's data

plt.pie(x=values, labels=names, autopct='%.1f%%') #shadow=True

plt.legend()
plt.title('The percentage of single nucluotide')

plt.show()
```

### [6]. Loss and Accuracy Plot:

```
def lossPlot(results):
    plt.title(label='Loss: Training and Validation')
    plt.plot(results.history['loss'], label='Training Loss')
    plt.plot(results.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
#end-def

def accuracyPlot(results):
    plt.title(label='Accuracy: Training and Validation')
    plt.plot(results.history['accuracy'], label='Training Accuracy')
    plt.plot(results.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
#end-def
```
